Exemplo prático do uso de regressão linear simples em Python


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# from yellowbrick.reressor import ResidualsPlot
```

```python
db = pd.read_csv('datasets\\cars.csv')
#colunas Unnamed, id, speed & dist
db.shape
db.head()
```

```python
db = db.drop(['Unnamed: 0'], axis=1)
db.head()
```

```python
X = db.iloc[:,1].values #dist - variavel independete
y = db.iloc[:, 0].values #speed - variavel dependente (depende da velocidade)
y
```

```python
correlacao = np.corrcoef(X, y)
correlacao #Quanto maior a correlação melhor a relaçao entre as variaveis
```

```python
X = X.reshape(-1,1)
model = LinearRegression()
model.fit(X,y)
```

```python
model.intercept_ #Coeficientes - coeficiente linear
```

```python
model.coef_ #inclinação da reta - coeficiente angular
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red')
```

```python
model.intercept_ + model.coef_ * 123
```

```python
model.predict([[123]])
```
